### Sampling distributions and confidence intervals

- When analysing data, we would like to estimate a sampling distribution in order to perform hypothesis tests and calculate confidence intervals.
- One way of attempting to solve this issue is a method called __bootstrapping__, whereby results for a population are inferred from repeated sampling.
- When generating samples using bootstrapping, this is done __with replacement__. This means that it is possible to ___select the same element from the sample more than once___.

```python
import scipy.stats as st
```

```python
x1 #petal.length
```

```python
np.random.seed(1234)
pd.DataFrame({'mean_of_means_of_Petal_length':[x1.sample(50, replace=True).mean() for i in range(0,10000)]})
```

```python
np.random.seed(1234)
bootstrap = pd.DataFrame({'mean_of_means_of_Petal_length':[x1.sample(50, replace=True).mean() for i in range(0,10000)]})
bootstrap
```

```python
bootstrap.hist()
plt.show()
```

```python
bootstrap.mean()
```

```python
bootstrap['mean_of_means_of_Petal_length'].quantile(0.025),bootstrap['mean_of_means_of_Petal_length'].quantile(0.975).round(2)
```

- Bootstrap 50 random observations of Petal length from the original dataset. Repeat this 10000 times. The mean of means of this bootstrapped sample is 3.76 with a confidence interval of (3.27, 4.25).
- However, this methods doesn't consider the reality of your population, e.g. different species of iris of your dataset. 
- As simply assuming that our dataset follows a normal distribution results in some error in your estimations, a bootstrap sample can only tell you things about the sample, and won't give you any new information about the real population. 
- If you want to gain more information about the population, you have to gather more data from the population.

## T-test

- (Research question) Are the difference in petal length between the species statistically different? 
- We can apply t-test to solve this problem. 

```python
from scipy import stats
import matplotlib.pyplot as plt
```

```python
iris.groupby("variety")['petal.length'].describe()
```

- Before the t-test, one needs to test its assumptions. First to test for __the homogeneity of variances__. To do this, I will use Levene’s test for homogeneity of variance. The method to conduct this test is stats.levene().

```python
stats.levene(Setosa['petal.length'], Versicolor['petal.length'])
# The small p-value suggests that the populations do not have equal variances. 
# This is not surprising, given that the sample variance of `Versicolor` is much larger than that of `Setosa`.
# We made the assumption of unequal variances between our two populations. 
# Welch’s t-test statistic does not assume that the population variances are equal and can be used whether the population variances are equal or not. 
# The test that assumes 'equal' population variances is referred to as the pooled t-test, aka Pooled Two-sampled t-test
# Pooling refers to finding a weighted average of the two independent sample variances.
```

- Second, before conducting the t-test, we also test the assumption of __normality__ using the stats.shapiro().

```python
stats.shapiro(Setosa['petal.length'])
# The first value in the tuple is the W test statistic, and the second value is the p-value.
```

```python
stats.shapiro(Versicolor['petal.length'])
# Neither of the variables of interest violates the assumption of normality so we can continue with our analysis plan. 
```

- Null hypothesis (H0): u1 = u2, which translates to the mean of sample 1 is equal to the mean of sample 2
- Alternative hypothesis (HA): u1 ≠ u2, which translates to the mean of sample 1 is not equal to the mean of sample 2

```python
stats.ttest_ind(Setosa['petal.length'], Versicolor['petal.length'], equal_var = False) #unequal variance
```

- The p-value is significant, therefore one can reject the null hypothesis in support of the alternative.
- In sum, Welch’s t-test was selected to analyze the data because Levene’s test for homogeneity of variances indicated unequal variances between groups. 
- The difference in petal length between the two species is significantly different.
