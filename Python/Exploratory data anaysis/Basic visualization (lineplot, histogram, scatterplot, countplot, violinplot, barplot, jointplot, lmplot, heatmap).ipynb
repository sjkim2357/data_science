## Line plot and scatter plot basics

## Plotting with exercise data

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
cars = pd.read_csv("cars.csv")
```

```python
cars
```

### Line plot

- __Create__ the line plot for all numerical columns.

```python
cars.plot(subplots = True, sharex = False, figsize = (15,15))
plt.show()
```

### Historam

- __Create__ the histogramm for the __"mpg"__ column with 40 bins.

```python
cars.mpg.hist(figsize = (12,8), bins = 40 )
plt.title("MPG", fontsize = 15)
plt.xlabel("mpg", fontsize = 13)
plt.ylabel("frequency", fontsize = 13)
plt.show()
```

### Scatter plot

- __Create__ the scatter plot with __horsepower__ and __mpg__.

```python
cars.plot(kind = "scatter", x = "horsepower", y = "mpg", figsize = (12,8), c = "cylinders", marker = "x", colormap = "viridis")
plt.title("MPG vs. Horsepower", fontsize = 18)
plt.xlabel("horsepower", fontsize = 15)
plt.ylabel("mpg", fontsize = 15)
plt.show()
```

# Advanced visualization with Seaborn:  Coding Exercises

## Import the packages and the dataset

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
cars = pd.read_csv("cars.csv")
```

## Countplot

- __sns.countplot()__ method is used to Show the counts of observations in each categorical bin using bars.

```python
plt.figure(figsize=(12,8))
sns.countplot(data = cars, x = "model_year", hue = "origin", palette= "viridis") #hue: column with categorical data
plt.show()
```

## Violinplot

- __sns.violinplot()__ method plays a similar role as a box plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. 

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.violinplot(data = cars, x = "origin", y = "mpg", dodge = True)
sns.swarmplot(data = cars, x = "origin", y = "mpg", dodge = True, color = "black")
plt.show()
```

## Barplot

- __sns.barplot()__ represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars. 

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.barplot(data = cars, x = "origin", y = "mpg", dodge = True)
plt.show()
```

## Pointplot

- __sns.poinplot()__ represents an estimate of central tendency for a numeric variable by the position of the dot and provides some indication of the uncertainty around that estimate using error bars.

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.pointplot(data = cars, x = "model_year", y = "mpg", hue = "origin", dodge = True)
plt.show()
```

## Jointplot

- __sns.jointplot()__ draws a plot of two variables with bivariate and univariate graphs.

```python
cars["gpm"] = (1/cars.mpg*100).round(2)
```

```python
cars.drop(columns = ["mpg"], inplace = True)
```

```python
cars
```

- __Create__ the plot that shows __relationship__ between horsepower and gpm.

```python
sns.set(font_scale=1.5)
sns.jointplot(data = cars, x = "horsepower", y = "gpm", height = 8, kind = "reg")
plt.show()
#regression
```

## Lmplot

- __sns.lmplot()__ method is used to draw a scatter plot onto a __FacetGrid__.
- __FacetGrid__ class helps in visualizing distribution of one variable as well as the relationship between multiple variables separately within subsets of your dataset using multiple panels.

```python
sns.set(font_scale=2)
sns.lmplot(data = cars, x = "weight", y = "horsepower", height=8, col = "origin")
plt.show()
#linear model for two subplots
```

## Heatmap

- __sns.heatmapt()__ is a plot of rectangular data as a color-encoded matrix. 
- This is a great way to visualize data, because it can show the relation between variabels. 

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.4)
sns.heatmap(cars.corr(), annot= True, cmap = "Reds") #correlation matrix
plt.show()
```
